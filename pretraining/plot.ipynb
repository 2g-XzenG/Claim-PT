```python
import matplotlib.pyplot as plt
import numpy
import _pickle as pickle
```

```python
def extract_line(lines, substr):
    loss = []
    for line in lines:
        if substr in line:
            l = line.partition(substr)[2].split(" ")[0]
            loss.append(float(l))
    return loss
```

```python
history_list = []
for i in [1,2,4,10]:
    
    with open("diff_head/" + str(i) + '.out') as f:
        lines = f.readlines()
        loss = extract_line(lines, "loss: ")
        x1, x2 = 1000,11000
        plt.plot(range(x1, x2), loss[x1:x2], label="# of head:"+str(i))
plt.legend()
plt.xlabel("# of iterations")
plt.ylabel("Loss")
plt.savefig("diff_head", dpi=600)
plt.show()
```
